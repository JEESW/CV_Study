```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagefile = '../1/lena.jpg'
image = cv2.imread(imagefile)
cv2.imshow('imread display', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(imageRGB)
plt.show()
```
