```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagepath = '../1/russianplate.jpg'
cascpath = '../xml/haarcascade_russian_plate_number.xml'
```

```python
ruscascade = cv2.CascadeClassifier(cascpath)

image = cv2.imread(imagepath)
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
faces = ruscascade.detectMultiScale(
    gray,                 #  
    minNeighbors = 5,     #    
    minSize = (20, 20)    #   
)

for (x, y, w, h) in faces:
    cv2.rectangle(imageRGB, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
plt.figure(figsize = (8, 8))
plt.axis('off')
plt.imshow(imageRGB)
```
